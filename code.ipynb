```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('appstore_games.csv')
df.describe()
```

```python
df=df.dropna(axis=0, subset=['URL', 'ID', 'Name', 'Icon URL', 'Average User Rating',
       'User Rating Count', 'Price', 'Description',
       'Developer', 'Age Rating', 'Languages', 'Size', 'Primary Genre',
       'Genres', 'Original Release Date', 'Current Version Release Date'])
```

```python
ndf=df.iloc[:,[2,5,6,7,8,15]]
ndf.head(10) #Extracting useful part of data, preparing data
```

```python
x=ndf['Average User Rating']
y=ndf['Price']
plt.scatter(x,y)
plt.ylabel('Price')
plt.xlabel('Average User Rating')
plt.show()
"""
This is a graphical representation of the trend between average user rating and price.
It shows that apps with higher user rating i.e, 4.0 and above have price in range 0 and $20.
"""
```

x=ndf['Price']
y=ndf['User Rating Count']
plt.scatter(x,y,color='Red')
plt.xlabel('Price')
plt.ylabel('User Rating Count')
plt.show()
"""
This is a graphical representation of the relation between Price and User Rating count.
It shows that the apps with lower price have maximum user rating count.
Which clearly shows that if we keep the price of an app low, maximum people will download it.
And also as per the previous scatter plot, maximum rating will come if we keep the price range in the range $0-$20
"""

```python
"""
    Listing the names of genres with higher user rating
"""
highestRating=ndf.loc[(ndf['Average User Rating']==5.0)]
appWithHigherRating=set()
for i in highestRating['Genres'].unique():
    for j in i.split(', '):
        appWithHigherRating.add(j)
appWithHigherRating.remove('Games')
print("Genres with higher rating")
count=1;
for i in appWithHigherRating:
    print(str(count)+". "+i)
    count+=1      
```
